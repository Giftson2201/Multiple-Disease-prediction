```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
import joblib
import warnings
warnings.filterwarnings('ignore')

print("Starting Phase 2: Medical Model Training...\n")
print("-" * 40)

# ==========================================
# MODEL 1: Parkinson's Disease
# ==========================================
try:
    print("1. Training Parkinson's Model...")
    parkinsons_data = pd.read_csv('parkinsons.csv')
    
    # Drop the 'name' column as it's not a medical feature
    if 'name' in parkinsons_data.columns:
        parkinsons_data = parkinsons_data.drop(columns=['name'], axis=1)
        
    X = parkinsons_data.drop(columns=['status'], axis=1) # 'status' is the target (1=disease, 0=normal)
    y = parkinsons_data['status']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
    
    # Scale the data
    scaler_p = StandardScaler()
    scaler_p.fit(X_train)
    X_train_scaled = scaler_p.transform(X_train)
    
    # Train model
    model_p = RandomForestClassifier(random_state=2)
    model_p.fit(X_train_scaled, y_train)
    
    # Save model and scaler
    joblib.dump(model_p, 'parkinsons_model.pkl')
    joblib.dump(scaler_p, 'parkinsons_scaler.pkl')
    print("✅ Parkinson's Model saved successfully!")

except Exception as e:
    print(f"❌ Error with Parkinson's: {e}")

print("-" * 40)

# ==========================================
# MODEL 2: Kidney Disease
# ==========================================
try:
    print("2. Training Kidney Disease Model...")
    kidney_data = pd.read_csv('kidney_disease.csv')
    
    # Drop ID column
    if 'id' in kidney_data.columns:
        kidney_data = kidney_data.drop('id', axis=1)
        
    # Handle missing values by dropping them for simplicity in this baseline model
    kidney_data = kidney_data.dropna()
    
    # Convert text columns to numbers
    le = LabelEncoder()
    for col in kidney_data.select_dtypes(include=['object']).columns:
        kidney_data[col] = le.fit_transform(kidney_data[col].astype(str))
        
    X = kidney_data.drop('classification', axis=1)
    y = kidney_data['classification']
    
    # In some datasets, CKD is encoded differently. Ensure 0 and 1.
    y = y.replace({'ckd': 1, 'notckd': 0, 'ckd\t': 1})
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
    
    model_k = RandomForestClassifier(random_state=2)
    model_k.fit(X_train, y_train)
    
    joblib.dump(model_k, 'kidney_model.pkl')
    # Save a list of columns so our Streamlit app knows what to expect
    joblib.dump(list(X.columns), 'kidney_columns.pkl') 
    print("✅ Kidney Disease Model saved successfully!")

except Exception as e:
    print(f"❌ Error with Kidney Disease: {e}")

print("-" * 40)

# ==========================================
# MODEL 3: Liver Disease
# ==========================================
try:
    print("3. Training Liver Disease Model...")
    liver_data = pd.read_csv('indian_liver_patient.csv')
    
    # Handle missing values
    liver_data['Albumin_and_Globulin_Ratio'].fillna(liver_data['Albumin_and_Globulin_Ratio'].median(), inplace=True)
    
    # Convert Gender to 0 and 1
    if 'Gender' in liver_data.columns:
        liver_data['Gender'] = liver_data['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
        
    X = liver_data.drop('Dataset', axis=1)
    y = liver_data['Dataset']
    
    # The dataset uses 1 for Liver Disease and 2 for No Disease. We map 2 to 0.
    y = y.map({1: 1, 2: 0})
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
    
    model_l = RandomForestClassifier(random_state=2)
    model_l.fit(X_train, y_train)
    
    joblib.dump(model_l, 'liver_model.pkl')
    print("✅ Liver Disease Model saved successfully!")

except Exception as e:
    print(f"❌ Error with Liver Disease: {e}")

print("-" * 40)
print("Phase 2 Complete! Check your folder for the new .pkl files.")
```
